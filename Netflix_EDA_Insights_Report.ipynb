# Step 2: Data Cleaning

```python
df['director'] = df['director'].fillna('Unknown')
```

```python
df['cast'] = df['cast'].fillna('Not Available')
```

```python
df['country'] = df['country'].fillna('Unknown')
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
df['duration'] = df['duration'].fillna('Unknown')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

## Cleaning Decisions

1. Missing director values were replaced with "Unknown".
2. Missing cast values were replaced with "Not Available".
3. Missing country values were replaced with "Unknown".
4. Missing ratings were replaced using the most frequent rating.
5. Duplicate rows were removed.
6. Date column was converted to datetime format.
7. Some invalid or missing values in the date_added column were converted to NaT using errors='coerce' during datetime conversion.

# Step 4: Data Visualization

### Chart 1: Movies vs TV Shows

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x='type')

plt.title("Movies vs TV Shows")
plt.xlabel("Content Type")
plt.ylabel("Count")

plt.show()
```

### Chart 2: Release Year Distribution

```python
plt.figure(figsize=(10,5))

plt.hist(df['release_year'], bins=20)

plt.title("Distribution of Release Years")
plt.xlabel("Release Year")
plt.ylabel("Frequency")

plt.show()
```

### Chart 3: Content Distribution

```python
df['type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Movies vs TV Shows Distribution")
plt.ylabel("")

plt.show()
```

### Chart 4: Netflix Content Growth

```python
year_data = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12,5))

plt.plot(year_data.index, year_data.values)

plt.title("Netflix Content Growth Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

### Chart 5: Release Year Distribution

```python
plt.figure(figsize=(10,5))

sns.scatterplot(
    data=df,
    x='release_year',
    y=df.index,
    hue='type'
)

plt.title("Release Year Distribution by Type")

plt.show()
```

### Visualization 6:Year-wise table

```python
year_type = pd.crosstab(
    df['release_year'],
    df['type']
)

plt.figure(figsize=(12,8))

sns.heatmap(
    year_type,
    cmap='YlGnBu'
)

plt.title("Heatmap of Release Year vs Content Type")
plt.xlabel("Content Type")
plt.ylabel("Release Year")

plt.show()
```

### Chart 7: Top 10 Countries Producing Netflix Content

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_countries.values,
    y=top_countries.index
)

plt.title("Top 10 Countries by Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")

plt.show()
```
